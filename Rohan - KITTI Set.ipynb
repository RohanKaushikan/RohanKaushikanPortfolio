```python
import numpy as np
import cv2, glob

color_dict = {'Tram':(255,0,0), 'Person_sitting':(0,255,0), 'Truck':(0,0,255),
              'DontCare':(255,255,0), 'Pedestrian':(0,255,255), 'Cyclist':(0,255,255),
              'Misc':(255,0,255), 'Van':(128,0,0), 'Car':(255,255,255)}
all_classes = set()
images = glob.glob('/content/gdrive/MyDrive/UCSC/*.png')
print(images)
#labels = glob.glob('./training/label_2/*.txt')

for img in images:
    image = cv2.imread(img)
    #breakpoint()
    number = img[-10:-4]
    label_path = '/content/gdrive/MyDrive/UCSC/training/label_2/' + number + '.txt'

    label_handle = open(label_path, 'r')
    lines = label_handle.readlines()
    label_handle.close()

    for line in lines:
        parsed = line.rstrip('\n').split(' ')

        class_name = parsed[0]
        bounding_box_2d = parsed[4:8]

        x0, y0, x1, y1 = float(bounding_box_2d[0]), float(bounding_box_2d[1]), float(bounding_box_2d[2]), float(bounding_box_2d[3])
        start_point = (int(x0), int(y0))
        end_point = (int(x1), int(y1))

        cv2.rectangle(image, start_point, end_point, color=color_dict[class_name], thickness=1)
        cv2.putText(image, class_name, (int(x0), int(y0-5)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color_dict[class_name], 1)
        all_classes.add(class_name)

    cv2.imwrite('./boundingboxes/' + number + '.png', image)
    print(img)
    #cv2.imshow('Rectangle', image)
    #cv2.waitKey(0)
    #breakpoint()

#breakpoint()
```
